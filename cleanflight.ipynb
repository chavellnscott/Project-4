```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf 
```

```python
flights_df = pd.read_csv('C:/Users/charl/Downloads/Atlanta_01_2023_Flights.csv')
```

```python
Atl_dept = flights_df[flights_df['ORIGIN_CITY_NAME'] == 'Atlanta, GA']
```

```python
Atl_dept
```

```python
AA_Atl_depts = Atl_dept[Atl_dept['MKT_UNIQUE_CARRIER'] == 'AA']
```

```python
AA_Atl_depts
```

```python
AA_Atl_depts = AA_Atl_depts.fillna(0)
```

```python
AA_Atl_depts
```

```python
AA_Atl_depts['Total_delay_minutes'] = AA_Atl_depts['DEP_DELAY_NEW'].where(AA_Atl_depts['DEP_DELAY_NEW'] != 0, 0).cumsum()
```

```python
AA_Atl_depts['Total_deptartures'] = AA_Atl_depts.groupby((AA_Atl_depts['DAY_OF_MONTH'] == 0).cumsum()).cumcount() + 1
```

```python
AA_Atl_depts['Total_delay_count'] = (AA_Atl_depts['DEP_DELAY_NEW'] != 0).cumsum()
```
